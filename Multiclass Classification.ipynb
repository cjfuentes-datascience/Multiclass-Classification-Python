```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

Explore what this dataset contains

```python
dir(digits)
```

```python
digits.data[0]
```

Let's plot the first five images

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

Let's train our model

```python
from sklearn.model_selection import train_test_split
```

```python
x_test, x_train, y_test, y_train = train_test_split(digits.data, digits.target, test_size=0.2)
```

Let's build or model

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train, y_train)
```

```python
log.score(x_test, y_test)
```

Let's pick a random sample

```python
y_predict = log.predict(x_test)
```

Let's get a feeling of how accurate our model is by using the 'confusion_matrix'

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predict)
cm
```

Let's visualize it in a heatmap

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
```
